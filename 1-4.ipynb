```python
#bfs
graph = {'A': ['B', 'D'], 'B': ['F', 'C','A'], 'C': ['E','G'], 'D': ['F'], 'F': ['A'],'E':['F'],'G':['E']}
my_q = ['A']
visited = []

while my_q:
    state = my_q.pop(0)
    visited.append(state)
    for neigbor in graph[state]:
        if neigbor not in visited and neigbor not in my_q:  #its its or
            my_q.append(neigbor)
            
        

print(visited)
```

```python
#dfs
graph = {'A': ['B', 'D'], 'B': ['F', 'C','A'], 'C': ['E','G'], 'D': ['F'], 'F': ['A'],'E':['F'],'G':['E']}
my_stk= ['A']
visited = []

while my_stk:
    state = my_stk.pop()
    visited.append(state)
    for neigbor in graph[state]:
        if neigbor not in visited and neigbor not in my_stk:
            my_stk.insert(0,neigbor)
            
        

print(visited)
```

```python
import random

board = [["." for i in range(3)] for _ in range(3)]

def print_board():
    for row in board:
        print('|'.join(row))
    print("------")

def check_win(player):
    for i in range(3):
        if all(board[i][j]== player for j in range(3)):
            return True
        
    for j in range(3):
        if all(board[i][j]== player for i in range(3)):
            return True
        
    # for i in range(3):
    #     if all(board[i][i] == player):
    #         return True
        
    # for i in range(3):
    #     if all(board[i][2-i] == player for i in range(3)):
    #         return True

    if all(board[i][i] == player for i in range(3)):
        return True
    
    if all(board[i][2-i] == player for i in range(3)):
        return True
    
    return False

def ai_movef_poss(player):
    possi = []
    for a in range(3):
        for b in range(3):
            if board[a][b] == ".":
                possi.append((a,b))
    if possi:
        move = random.choice(possi)
        return move
    else:
        return False

turn = 0

while True:
    print_board()
    candidates = ['x','y']
    
    player = candidates[turn%2]
    print(f"player is {player}")
    move = ai_movef_poss(player)
    
    if move:
        board[move[0]][move[1]] = player
    else:
        print("its a tie, no moves left")
        break

    if check_win(player):
        print(f"{player}  won the game")
        break

    turn+=1

print_board()
    
    
        
```

```python
capacity_jug1 = 4
capacity_jug2 = 3

goal = 2

def is_goal(state):
    return goal in state

def generate_neighbours(state):
    j1 = state[0]
    j2 = state[1]

  
    ngbr = []

    if j1 < capacity_jug1:
        ngbr.append((capacity_jug1,j2))

    if j2 < capacity_jug2:
        ngbr.append((j1,capacity_jug2))

    if j1 != 0:
        ngbr.append((0,j2))

    if j2 != 0:
        ngbr.append((j1,0))

    #transfer from jug1 to jug2

    t = min(j1,capacity_jug2 - j2)
    ngbr.append((j1-t,j2+t))

#cpacity to transfer from jug2 to j1 
    t = min(capacity_jug1 - j1,j2)
    ngbr.append((j1+t, j2-t))

    return ngbr



def bfs():
    visited = []
    start = (0,0)

    q = [(start,[])]

    while q:
        node = q.pop(0)
        state = node[0]
       
        path= node[1]

        if is_goal(state):
            return path + [state]
        
        visited.append(state)
        
        for ngb in generate_neighbours(state):
            if ngb not in visited or ngb not in [k for k,v in q] :
                new_path = path + [state]
                q.append((ngb,new_path))





sol = bfs()

if sol:
    print("solution:: -->")
    for step in sol:
        print(step)




```
