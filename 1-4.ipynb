```python
#bfs
graph = {'A': ['B', 'D'], 'B': ['F', 'C','A'], 'C': ['E','G'], 'D': ['F'], 'F': ['A'],'E':['F'],'G':['E']}
my_q = ['A']
visited = []

while my_q:
    state = my_q.pop(0)
    visited.append(state)
    for neigbor in graph[state]:
        if neigbor not in visited and neigbor not in my_q:
            my_q.append(neigbor)
            
        

print(visited)
```

```python
#dfs
graph = {'A': ['B', 'D'], 'B': ['F', 'C','A'], 'C': ['E','G'], 'D': ['F'], 'F': ['A'],'E':['F'],'G':['E']}
my_stk= ['A']
visited = []

while my_stk:
    state = my_stk.pop()
    visited.append(state)
    for neigbor in graph[state]:
        if neigbor not in visited and neigbor not in my_stk:
            my_stk.insert(0,neigbor)
            
        

print(visited)
```

```python
import random

board = [["." for i in range(3)] for _ in range(3)]

def print_board():
    for row in board:
        print('|'.join(row))
    print("------")

def check_win(player):
    for i in range(3):
        if all(board[i][j]== player for j in range(3)):
            return True
        
    for j in range(3):
        if all(board[i][j]== player for i in range(3)):
            return True
        
    # for i in range(3):
    #     if all(board[i][i] == player):
    #         return True
        
    # for i in range(3):
    #     if all(board[i][2-i] == player for i in range(3)):
    #         return True

    if all(board[i][i] == player for i in range(3)):
        return True
    
    if all(board[i][2-i] == player for i in range(3)):
        return True
    
    return False

def ai_movef_poss(player):
    possi = []
    for a in range(3):
        for b in range(3):
            if board[a][b] == ".":
                possi.append((a,b))
    if possi:
        move = random.choice(possi)
        return move
    else:
        return False

turn = 0

while True:
    print_board()
    candidates = ['x','y']
    
    player = candidates[turn%2]
    print(f"player is {player}")
    move = ai_movef_poss(player)
    
    if move:
        board[move[0]][move[1]] = player
    else:
        print("its a tie, no moves left")
        break

    if check_win(player):
        print(f"{player}  won the game")
        break

    turn+=1

print_board()
    
    
        
```
